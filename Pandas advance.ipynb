# Pandas Merging Function

```python
import pandas as pd
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/attaurrahman294/Datasets/main/student_result1.csv')
df
```

```python
df['New'] = ['Sajid' , 'Salma' , 'Iqrar' ,'Mis' , 'Wahab']
df2 = df.drop(['Section' , 'Class' , 'Percentage'] , axis=1)
```

```python
df2
```

```python
df = df.drop(['Section' , 'Class' , 'Percentage' , 'New'] , axis=1)
```

```python
 pd.merge(df,df2 , on=['Student ID'])
```

```python
 pd.merge(df,df2 , on=['Student ID' , 'Study hrs' , 'Social Media usage hrs'])
```

```python
df2.loc[4 , 'Student ID'] = 1002
df2
```

```python
df2
```

```python
# this merge are doing intersection of dataframe onle take the same records and descard the differet
pd.merge(df,df2 , on=['Student ID' , 'Study hrs' , 'Social Media usage hrs'])
```

```python
# this also take the intersection
pd.merge(df,df2, on=['Student ID' , 'Study hrs' , 'Social Media usage hrs'])
```

```python
# we can see the both have gettign intersection
pd.merge(df,df2 , on=['Student ID' , 'Study hrs' , 'Social Media usage hrs'] ,how='inner' , indicator=True)
```

```python
# we can see the both have gettign intersection
pd.merge(df,df2 , on=['Student ID' , 'Study hrs' , 'Social Media usage hrs'] ,how='outer' , indicator=True)
```

```python
# this will retrive only the left column column and the data of write only whecn have same Student ID
pd.merge(df,df2, on = ['Student ID' , 'Study hrs' , 'Social Media usage hrs'] ,how='left' , indicator=True)
#'left', 'right', 'outer', 'inner', 'cross'
```

```python
# this will retrive all the Std id key having same or differ
pd.merge(df,df2, on = ['Student ID' , 'Study hrs' , 'Social Media usage hrs'] ,how='outer' , indicator=True)
#'left', 'right', 'outer', 'inner', 'cross'
```

```python
pd.merge(df,df2, on = ['Student ID' , 'Study hrs' , 'Social Media usage hrs'] ,how='outer' , indicator=True)
#'left', 'right', 'outer', 'inner', 'cross'
```

```python
# this will retrive the data having same same id and all the id of right 
pd.merge(df,df2, on = ['Student ID' , 'Study hrs' , 'Social Media usage hrs'] ,how='right' , indicator=True)
#'left', 'right', 'outer', 'inner', 'cross'
```
